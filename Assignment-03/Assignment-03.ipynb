# Boston House Price

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression, Ridge, Lasso
```

```python
boston = load_boston()
boston
```

```python
X = boston.data
y = boston.target
```

```python
reg = LinearRegression()
#lasso_reg = Lasso(alpha=5)
ridge_reg = Ridge(alpha=5)

reg.fit(X,y)
#lasso_reg.fit(X,y)
ridge_reg.fit(X,y)
```

```python
for a, b in zip(boston.feature_names, reg.coef_):
    print("%s: %f" %(a, b))
```

```python
reg.intercept_
```

```python
#for a, b in zip(boston.feature_names, lasse_reg.coef_):
for a, b in zip(boston.feature_names, ridge_reg.coef_):
    print("%s: %f" %(a, b))
```

```python
reg.score(X, y)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 32))
for n in range(13):
    x = [i[n] for i in X]
    
    plt.subplot(7, 2, n+1)
    plt.title(boston.feature_names[n])
    plt.scatter(x, y)
```

### Report

#### 이 데이터로는 약 74%의 경우만 예측 가능성이 있고
#### 각 요인과 Boston Houes Price의 상관관계는 위의 차트를 통해 엿볼 수 있겠다.
